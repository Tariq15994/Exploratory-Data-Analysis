```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("Data_Train.xlsx")
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df=train_df.append(test_df)
final_df.head()
```

```python
final_df.info()
```

```python
final_df["Day"] = final_df["Date_of_Journey"].apply(lambda x : x.split('/')[0])
```

```python
final_df["Month"] = final_df["Date_of_Journey"].apply(lambda x: x.split("/")[1])
final_df["Year"] = final_df["Date_of_Journey"].apply(lambda x: x.split("/")[2])
```

```python
final_df.head()
```

```python
final_df["Day"]= final_df["Day"].astype(int)
final_df["Year"]= final_df["Year"].astype(int)
final_df["Month"]= final_df["Month"].astype(int)
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df["Arrival_Time"] = final_df["Arrival_Time"].apply(lambda x: x.split(' ')[0])
```

```python
final_df["Arrival_hour"] =  final_df["Arrival_Time"].apply(lambda x: x.split(":")[0])
final_df["Arrival_min"] =  final_df["Arrival_Time"].apply(lambda x: x.split(":")[1])
```

```python
final_df["Arrival_hour"]= final_df["Arrival_hour"].astype(int)
final_df["Arrival_min"]= final_df["Arrival_min"].astype(int)
```

```python
final_df.drop("Arrival_Time",axis=1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df["Dep_hour"] = final_df["Dep_Time"].apply(lambda x: x.split(' ')[0]).apply(lambda x: x.split(":")[0])
final_df["Dep_min"] = final_df["Dep_Time"].apply(lambda x: x.split(' ')[0]).apply(lambda x: x.split(":")[1])

final_df["Dep_hour"]= final_df["Dep_hour"].astype(int)
final_df["Dep_min"]= final_df["Dep_min"].astype(int)
final_df.drop("Dep_Time",axis=1,inplace=True)
```

```python
#let's see unique vlaue
final_df["Total_Stops"].unique() # there is nan value 
```

```python
#how many nan value are there
final_df["Total_Stops"].isnull().sum()
```

```python
final_df.head()
```

```python
# converting Total_stops to int 
final_df["Total_Stops"] = final_df["Total_Stops"].map({'non-stop':0,'1 stop':2,'2 stops':2,'3 stops':3,"NaN":1})
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
#convert duration into mintes
final_df["Duration"].apply(lambda x: x.split(' ')[0]).apply(lambda x: x.split("h")).str[0].astype(int)
```

```python
# Error is somewhere in the dataset the 5m value is there so first we need to solve this error 
# let's check where is 5m 

final_df[final_df['Duration'] == '5m']
# so as we know we duration of mumbai to hyderabad couldn't be 5mintes, we should drop these record
```

```python
final_df.drop([6474,2660],axis =0,inplace=True)
```

```python
#convert duration into mintes
final_df["Duration_hour"] = final_df["Duration"].apply(lambda x: x.split(' ')[0]).apply(lambda x: x.split("h")).str[0].astype(int)*60
# final_df["Duration"].apply(lambda x: x.split(' ')[1]).apply(lambda x: x.split("m")).str[0].astype(int)
```

```python
final_df["Duration_min"] = final_df["Duration"].apply(lambda x: x.split('h')[1]).apply(lambda x:x.split('m')[0])
```
